```python
def add(a,b):
    return(a+b)
def sub(a,b):
    return(a-b)
def mul(a,b):
    return(a*b)
def div(a,b):
    return(a/b)
while True:
    d=int(input("enter a number : "))
    e=int(input("enter a number : "))
    print("1.add",end=" ")
    print("2.substract",end=" ")
    print("3.multiply",end=" ")
    print("4.division",end=" ")
    print("5.exit",end=" ")
    print(" ")
    c=int(input("enter choice "))
    if c==1:
        print("addition is : ",add(d,e))
    elif c==2:    
        print("substraction is : ",sub(d,c))
    elif c==3:
        print("multiplacation is : ",mul(d,e))
    elif c==4:
        print("addition is : ",div(d,e))
    elif c==5:
        print("goodbye")
        break
    else:
        print("enter correct choice")
        
```
